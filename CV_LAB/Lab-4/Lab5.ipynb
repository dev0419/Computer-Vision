```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt
```

Canny Edge Detection

```python
image = cv2.imread('justice.jpg',cv2.IMREAD_GRAYSCALE)
edges = cv2.Canny(image,100,200)
plt.subplot(1,2,1)
plt.imshow(image,cmap='gray')
plt.title('Original Image')
plt.axis('off')
plt.subplot(1,2,2)
plt.imshow(edges,cmap='gray')
plt.title('Canny Edge')
plt.axis('off')
```

Hough Transform

```python
img = cv2.imread('sudoku.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
edges = cv2.Canny(gray, 50, 150, apertureSize=3)
lines = cv2.HoughLines(edges, 1, np.pi / 180, 150)

for line in lines:
    rho, theta = line[0]
    a = np.cos(theta)
    b = np.sin(theta)
    x0 = a * rho #provides origin/top-left corner of the image
    y0 = b * rho #provides origin/top-left corner of the image
    x1 = int(x0 + 1000 * (-b)) #to get line coordinates
    y1 = int(y0 + 1000 * (a))
    x2 = int(x0 - 1000 * (-b))
    y2 = int(y0 - 1000 * (a))
    cv2.line(img, (x1, y1), (x2, y2), (0, 0, 255), 2)

# Display the image using Matplotlib
plt.subplot(2,2,1)
plt.imshow(gray,cmap='gray')
plt.title('original image')
plt.axis('off')

plt.subplot(2,2,3)
plt.imshow(edges,cmap='gray')
plt.title('Canny Edge Detected')
plt.axis('off')

plt.subplot(2,2,2)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Line Detected')
plt.axis('off')
plt.show()

```

Linear Filtering 

```python
image = cv2.imread('justice.jpg')
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
rows,cols,_ = image.shape
noise = np.random.normal(0,15,(rows,cols,3)).astype(np.uint8)
noise_img = image + noise
plt.figure(figsize=(5,3))
plt.axis('off')
plt.imshow(noise_img)
plt.show()
```

```python
kernel = np.ones((6,6))/36
filtered_image = cv2.filter2D(src=noise_img,ddepth=-1,kernel=kernel)
plt.subplot(1,2,1)
plt.imshow(filtered_image)
plt.title('Filtered Image')
plt.axis('off')
plt.subplot(1,2,2)
plt.imshow(noise_img)
plt.title('Noise Image')
plt.axis('off')
plt.show()
```

```python
kernel = np.ones((4,4))/16
image_filtered = cv2.filter2D(src=noise_img,ddepth=-1,kernel=kernel)
plt.subplot(1,2,1)
plt.imshow(image_filtered)
plt.title('filtered_image')
plt.axis('off')
plt.subplot(1,2,2)
plt.imshow(noise_img)
plt.title('noise_image')
plt.axis('off')
plt.show()
```

Gaussian blur 

```python
image_filtered = cv2.GaussianBlur(noise_img,(5,5),sigmaX=4,sigmaY=4)
plt.subplot(1,2,1)
plt.imshow(image_filtered)
plt.title('filtered_image')
plt.axis('off')
plt.subplot(1,2,2)
plt.imshow(noise_img)
plt.title('noise_image')
plt.axis('off')
plt.show()
```

```python
image_filtered = cv2.GaussianBlur(noise_img,(11,11),sigmaX=10,sigmaY=10)
plt.subplot(1,2,1)
plt.imshow(image_filtered)
plt.title('filtered_image')
plt.axis('off')
plt.subplot(1,2,2)
plt.imshow(noise_img)
plt.title('noise_image')
plt.axis('off')
plt.show()
```

```python
kernel = np.array([[-1,-1,-1],[-1,9,-1],[-1,-1,-1]])
sharpened = cv2.filter2D(image,-1,kernel)
plt.subplot(1,2,1)
plt.imshow(image)
plt.title('Original Image')
plt.axis('off')
plt.subplot(1,2,2)
plt.imshow(sharpened)
plt.title('Sharpened Image')
plt.axis('off')
plt.show()
```
